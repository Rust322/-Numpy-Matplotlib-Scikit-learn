```python
t = np.linspace(0, 10, 51)
t
```

```python
f = np.cos(t)
f
```

```python
plt.plot(t,f, color = 'green')
plt.show
```

```python
#3
plt.plot(t,f, color = 'green')
plt.title(' f(t)')
plt.xlabel(' t')
plt.ylabel(' f')
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.show
```

```python
x = np.linspace(-3, 3, 51)
x
```

```python
y1 = x**2
y2 = 2*x + 0.5
y3 = -3*x - 1.5
y4 = np.sin(x)
```

```python
fig = plt.figure()
fig, ax = plt.subplots(nrows=2, ncols=2)

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace = 0.3, hspace = 0.3)

ax[0, 0].plot(x, y1)
ax[0, 0].set(title = ' y1', xlim = (-5, 5))
ax[0, 1].plot(x, y2)
ax[0, 1].set_title(' y2')
ax[1, 0].plot(x, y3)
ax[1, 0].set_title(' y3')
ax[1, 1].plot(x, y4)
ax[1, 1].set_title(' y4')
plt.show
```
